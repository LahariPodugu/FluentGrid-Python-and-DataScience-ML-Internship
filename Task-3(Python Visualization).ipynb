```python
#1)Below graph shows the overall Visa application distribution status
```

```python
#Import package
import matplotlib.pyplot as plt
import seaborn as sns

# Count the number of applications per case status
case_status_counts = df["case_status"].value_counts()
print(case_status_counts)

# Plot the distribution of visa application status
plt.figure(figsize=(8, 5))
sns.barplot(x=case_status_counts.index, 
            y=case_status_counts.values, 
            hue=case_status_counts.index,  
            legend=False, 
            palette="viridis")

# Add labels and title
plt.title("Visa Application Status Distribution", fontsize=14)
plt.xlabel("Case Status", fontsize=12)
plt.ylabel("Number of Applications", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

```python
#2)Below graph shows the visa application distribution by continents
```

```python
# Count the number of applications per continent
continent_counts = df["continent"].value_counts()
print(continent_counts)

# Plot the distribution of visa applications by continent
plt.figure(figsize=(10, 6))
sns.barplot(x=continent_counts.index, 
            y=continent_counts.values, 
            hue=continent_counts.index,  
            legend=False,  
            palette="coolwarm")

# Add labels and title
plt.title("Number of Visa Applications by Continent", fontsize=14)
plt.xlabel("Continent", fontsize=12)
plt.ylabel("Number of Applications", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

```python
#3)Below graph shows the Prevailing Wage Distribution by Visa Case Status
```

```python
# Plot a box plot of prevailing wage by case status
plt.figure(figsize=(10, 6))
sns.boxplot(x="case_status", 
            y="prevailing_wage", 
            data=df, 
            hue="case_status",  
            legend=False,  
            palette="Set2")

# Add labels and title
plt.title("Prevailing Wage Distribution by Visa Case Status", fontsize=14)
plt.xlabel("Case Status", fontsize=12)
plt.ylabel("Prevailing Wage", fontsize=12)
plt.yscale("log")  # Log scale to handle outliers
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

```python
#4)Below graph shows the Visa Application Status by Education Level
```

```python
# Count visa status by education level
education_status_counts = df.groupby(["education_of_employee", "case_status"]).size().unstack()
print(education_status_counts)

# Plot a stacked bar chart
education_status_counts.plot(kind="bar", stacked=True, figsize=(12, 6), colormap="viridis")

# Add labels and title
plt.title("Visa Application Status by Education Level", fontsize=14)
plt.xlabel("Education Level", fontsize=12)
plt.ylabel("Number of Applications", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Case Status")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

```python
#5)Below graph shows the Trend of Visa Applications Over the Years
```

```python
# Count applications by year of establishment
yearly_counts = df["yr_of_estab"].value_counts().sort_index()
print(yearly_counts)

# Plot the trend of visa applications over time
plt.figure(figsize=(12, 6))
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values, marker="o", color="b")

# Labels and title
plt.title("Trend of Visa Applications Over the Years", fontsize=14)
plt.xlabel("Year of Establishment", fontsize=12)
plt.ylabel("Number of Applications", fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```

```python
#6)Below graph shows the Visa Applications by Region of Employment
```

```python
# Count applications by region
region_counts = df["region_of_employment"].value_counts()
print(region_counts)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(region_counts, labels=region_counts.index, autopct="%1.1f%%", colors=["#ff9999", "#66b3ff", "#99ff99", "#ffcc99", "#c2c2f0"], startangle=140)

# Add title
plt.title("Visa Applications by Region of Employment", fontsize=14)

# Show the plot
plt.show()
```

```python
#7)Below graph shows the Impact of Job Experience on Visa Approval
```

```python
# Count visa case status by job experience
experience_status_counts = df.groupby(["has_job_experience", "case_status"]).size().unstack()
print(experience_status_counts)

# Plot grouped bar chart
experience_status_counts.plot(kind="bar", figsize=(8, 5), colormap="coolwarm")

# Labels and title
plt.title("Impact of Job Experience on Visa Approval", fontsize=14)
plt.xlabel("Has Job Experience", fontsize=12)
plt.ylabel("Number of Applications", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Case Status")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

```python
#8)Below graph shows the Distribution of Prevailing Wages
```

```python
# Plot the histogram for prevailing wages
plt.figure(figsize=(10, 6))
sns.histplot(df["prevailing_wage"], bins=50, kde=True, color="purple")

# Labels and title
plt.title("Distribution of Prevailing Wages", fontsize=14)
plt.xlabel("Prevailing Wage", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xscale("log")  # Log scale to handle large variations in salary
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

```python
#9)Below graph shows the Impact of Full-Time vs. Part-Time Positions on Visa Approvals
```

```python
# Count visa case status by full-time or part-time position
fulltime_status_counts = df.groupby(["full_time_position", "case_status"]).size().unstack()
print(fulltime_status_counts)

# Plot grouped bar chart
fulltime_status_counts.plot(kind="bar", figsize=(8, 5), colormap="coolwarm")

# Labels and title
plt.title("Impact of Full-Time vs. Part-Time Positions on Visa Approvals", fontsize=14)
plt.xlabel("Full-Time Position", fontsize=12)
plt.ylabel("Number of Applications", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Case Status")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

```python
#10)Below graph shows the Visa Approval Rates by Continent
```

```python
# Count visa case status by continent
continent_status_counts = df.groupby(["continent", "case_status"]).size().unstack()
print(continent_status_counts)

# Plot stacked bar chart
continent_status_counts.plot(kind="bar", stacked=True, figsize=(10, 6), colormap="viridis")

# Labels and title
plt.title("Visa Approval Rates by Continent", fontsize=14)
plt.xlabel("Continent", fontsize=12)
plt.ylabel("Number of Applications", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Case Status")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```
